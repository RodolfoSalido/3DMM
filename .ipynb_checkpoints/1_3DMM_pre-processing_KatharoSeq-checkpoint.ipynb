## KatharoSeq Functions

```python
# define the allosteric sigmoid equation
def allosteric_sigmoid(x, h, k_prime):
    y = x ** h / (k_prime + x ** h)
    return y

#Define curve fitting function
def katharoseq_fit_curve(input_table_df,filter_out=None):
    katharo_table_df = input_table_df[input_table_df.index.str.contains('KATHARO')].copy()
    # Calculate the total number of reads per sample
    katharo_table_df['deblur_reads'] = katharo_table_df.sum(axis=1)
    
    # calculate the number of reads aligning to the mock community input genera
    katharo_table_df['kath_reads'] = katharo_table_df[['k__Bacteria;p__Proteobacteria;c__Betaproteobacteria;o__Burkholderiales;f__Comamonadaceae;g__Variovorax']].sum(1)
    # calculate the percent correctly assigned
    katharo_table_df['correct_assign'] = katharo_table_df['kath_reads'] / katharo_table_df['deblur_reads']
    # subset out KatharoSeq columns and add in logarithim of reads for plotting
    katharo = katharo_table_df[['correct_assign','deblur_reads','kath_reads']].copy()
    katharo['log_deblur_reads'] = np.log10(katharo['deblur_reads'])
    if filter_out != None:
        # Filter out list of problematic samples
        katharo = katharo[~katharo.index.isin(filter_out)]
    #fit allosteric sigmoid
    popt, pcov = curve_fit(allosteric_sigmoid, katharo['log_deblur_reads'], katharo['correct_assign'], method='dogbox')
    
    return katharo, popt, pcov

#Define function that regresses a min read_threshold from a proportion of fidelity (from control samples)
def katharo_read_threshold(popt,proportion):
    # Determine the number of reads at which the specified proportion of reads aligning to the control reference

    # assign variables and solve for X (number of reads to pass filter)
    h = popt[0]  # first value printed above graph
    k = popt[1]   # second value printed above graph
    y = proportion ## what you want to solve for

    min_log_reads = np.power((k/((1/y)-1)),(1/h))
    min_freq = np.power(10, min_log_reads).astype(int)
    return min_freq
```

# Katharoseq filtering on min_reads for 0.80 alignment to target. 

First performs KatharoSeq analysis on a per_sequence_run basis, filters per_sequence_run feature_tables based on a read threshold that matches a 0.80 fidelity to the expected taxon from control samples, as modeled by the allosteric sigmoid fit.

Then performs a second KatharoSeq analysis on a per_extraction_plate basis, filters per_extraction_plate feature_tables based on a read threshold that matches a 0.80 fidelity to the expected taxon from control samples, as modeled by the allosteric sigmoid fit.

Lastly plots the results with matplotlib subplots. 

```python
##setup subplots, 12 curves so 4 columns by 3 rows for a landscape view
index=0
rows=3
columns=4
f, axes = plt.subplots(nrows=rows, ncols=columns, figsize=(20,10),sharey=True,sharex=True)
plt.subplots_adjust(hspace=0.3)

#per_sequence_run loop
for seq_run in metadata_df['run_prefix'].unique():
    f_table = filter_samples(collapsed_table,
                                     metadata = metadata,
                                     where= f"run_prefix == '{seq_run}'")

    filtered_table_df = f_table.filtered_table.view(pd.DataFrame)
    katharo, popt, pcov = katharoseq_fit_curve(filtered_table_df)
    
    # plot fit curve
    x = np.linspace(0, 6, 50)
    y = allosteric_sigmoid(x, *popt)

    # multi-plot the fit
    axes[index//columns,index%columns].plot(katharo['log_deblur_reads'], katharo['correct_assign'], 'o', label='data')
    axes[index//columns,index%columns].plot(x,y, label='fit')
    axes[index//columns,index%columns].set_ylim([0,1.05])
    #axes[index//columns,index%columns].legend(loc='best')
    axes[index//columns,index%columns].set_title(seq_run.replace('_S1_L001',""))
    
    
    axes[index//columns,index%columns].axvline(np.log10(katharo_read_threshold(popt,0.80)),ls='--')
    axes[index//columns,index%columns].axhline(0.80,ls='--')
    index +=1
    
#per_extraction_plate loop
for sample_plate in metadata_df['sample_plate'].unique():
    f_table = filter_samples(collapsed_table,
                                     metadata = metadata,
                                     where= f"sample_plate == '{sample_plate}'")
    plate_name = sample_plate.replace('3DMM_20220614_14542_','').replace('3DMM_20220623_14542_','')
#     f_table.filtered_table.save('./data/per_plate/' + '150353_collapsed_table_' + plate_name+'.qza')
    filtered_table_df = f_table.filtered_table.view(pd.DataFrame)
    katharo, popt, pcov = katharoseq_fit_curve(filtered_table_df)
    
    # plot fit curve
    x = np.linspace(0, 6, 50)
    y = allosteric_sigmoid(x, *popt)

    # plot the fit
    axes[index//columns,index%columns].plot(katharo['log_deblur_reads'], katharo['correct_assign'], 'o', label='data')
    axes[index//columns,index%columns].plot(x,y, label='fit')
    axes[index//columns,index%columns].set_ylim([0,1.05])
    #axes[index//columns,index%columns].legend(loc='best')
    axes[index//columns,index%columns].set_title(plate_name + ' read_threshold: '
                                                 + str(katharo_read_threshold(popt,0.80)))
    
    axes[index//columns,index%columns].axvline(np.log10(katharo_read_threshold(popt,0.80)),ls='--')
    axes[index//columns,index%columns].axhline(0.80,ls='--')
    index +=1
    
```

The KatharoSeq analysis was rerun after fitlering outliers from Plate_2 and Plate_4 to improve allosteric sigmoidal curve fit. Rerunning will overwrite the files written on the cells above. 

Since the curve fit of sequencing run '3DMM_Plates_1-5' was poor (didn't generalize well, likely due to differential sequencing coverage across extraction_plates), a per_extraction_plate KatharoSeq analysis was chosen. 

The KatharoSeq analysis was run on per_extraction_plate basis, with a feature_table collapsed at the genera level (collapsed_table),but we want to keep the higher resolution sOTUs featues from deblur instead, so we filter deblur_tables (instead of collapsed_table) with the KatharoSeq minimum sequencing depth threshold and saved to file (lines 66-69). 

```python
problematic_samples = ['150297.14542.KATHARO.P4.2','150297.14542.KATHARO.P2.4']
##setup subplots, 12 curves so 4 columns by 3 rows for a landscape view
index=0
rows=3
columns=4
f, axes = plt.subplots(nrows=rows, ncols=columns, figsize=(12,10),sharey=True,sharex=True)
plt.subplots_adjust(hspace=0.3)

#per_sequence_run loop
for seq_run in metadata_df['run_prefix'].unique():
    f_table = filter_samples(collapsed_table,
                                     metadata = metadata,
                                     where= f"run_prefix == '{seq_run}'")
    filtered_table_df = f_table.filtered_table.view(pd.DataFrame)
    katharo, popt, pcov = katharoseq_fit_curve(filtered_table_df)
    
    # plot fit curve
    x = np.linspace(0, 6, 50)
    y = allosteric_sigmoid(x, *popt)

    # plot the fit
    axes[index//columns,index%columns].plot(katharo['log_deblur_reads'], katharo['correct_assign'], 'o', label='data')
    axes[index//columns,index%columns].plot(x,y, label='fit')
    axes[index//columns,index%columns].set_ylim([0,1.05])
    axes[index//columns,index%columns].set_title(seq_run.replace('_S1_L001',""))
    axes[index//columns,index%columns].set_ylabel('Proportion of Reads on Target')
    axes[index//columns,index%columns].set_xlabel('Log10 (Read Counts)')
    
    #Plot thresholding lines
    axes[index//columns,index%columns].axvline(np.log10(katharo_read_threshold(popt,0.80)),ls='--')
    axes[index//columns,index%columns].axhline(0.80,ls='--')
    
    
    index +=1
    
#per_extraction_plate loop
for sample_plate in metadata_df['sample_plate'].unique():
    f_table = filter_samples(collapsed_table,
                                     metadata = metadata,
                                     where= f"sample_plate == '{sample_plate}'")
    f_deblur = filter_samples(deblur_table,
                                     metadata = metadata,
                                     where= f"sample_plate == '{sample_plate}'")
    plate_name = sample_plate.replace('3DMM_20220614_14542_','').replace('3DMM_20220623_14542_','')
    filtered_table_df = f_table.filtered_table.view(pd.DataFrame)
    katharo, popt, pcov = katharoseq_fit_curve(filtered_table_df,filter_out = problematic_samples)
    
    # plot fit curve
    x = np.linspace(0, 6, 50)
    y = allosteric_sigmoid(x, *popt)

    # plot the fit
    axes[index//columns,index%columns].plot(katharo['log_deblur_reads'], katharo['correct_assign'], 'o', label='data')
    axes[index//columns,index%columns].plot(x,y, label='fit')
    axes[index//columns,index%columns].set_ylim([0,1.05])
    katharo_min_read = katharo_read_threshold(popt,0.80)
    axes[index//columns,index%columns].set_title(plate_name + ' read_threshold: '
                                                 + str(katharo_min_read))
    axes[index//columns,index%columns].set_ylabel('Proportion of Reads on Target')
    axes[index//columns,index%columns].set_xlabel('Log10 (Read Counts)')
    
    #plot thresholding lines
    axes[index//columns,index%columns].axvline(np.log10(katharo_read_threshold(popt,0.80)),ls='--')
    axes[index//columns,index%columns].axhline(0.80,ls='--')
    
    # apply katharoseq min_read_threshold per plate to deblur table. 
    katharo_filtered_plate = filter_samples(f_deblur.filtered_table,min_frequency=katharo_min_read)
    katharo_filtered_plate.filtered_table.save('./data/katharo_filtered_deblur/151557_feature-table_' +
                                               plate_name + '.qza')
    index +=1
plt.tight_layout()
```
